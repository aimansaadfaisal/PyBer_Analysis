```python
#Import Dependencies
import matplotlib .pyplot as plt 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Using an Object Oriented approach create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis) 
plt.show()##
```

```python
#Create a plot and add a lable for the legend
plt.plot(x_axis, y_axis, color="blue", linewidth=2, label="Boston")
#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create the y axis limit
plt.ylim(0,45)
#Create a Title
plt.title("PyBer Fare By Month")
#Create a grid
plt.grid()
#Add a legend
plt.legend
```

```python
#Creating Bar charts using Matlab Method
#Create the plot
plt.bar(x_axis,y_axis)
```

```python
#Create a plot
plt.bar(x_axis,y_axis,color="green",label="Boston")
#Create labels for x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create title
plt.title("Pyber Fare By Month")
#Add the legend
plt.legend()
```

```python
#Create Scatter Plot using ,Matlab method
plt.plot(x_axis,y_axis,'o')
plt.scatter(x_axis,y_axis,label="Chicago")
#Create labels for x and y axis
plt.ylabel("Date")
plt.xlabel("Fare($)")
#Create title
plt.title("Pyber Fare By Month")
#Add the legend
plt.legend()
```

```python
#Create Scatter Plot using ,Matlab method
plt.plot(x_axis,y_axis,'o')
plt.scatter(x_axis,y_axis,s=y_axis,label="Chicago")
#Create labels for x and y axis
plt.ylabel("Date")
plt.xlabel("Fare($)")
#Create title
plt.title("Pyber Fare By Month")
#Add the legend
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Creating a scatter plot using Python list comprehension
plt.plot(x_axis,y_axis,'o')
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
#Creating scatter plot using object oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#Creating bubble graph using object oriented interface
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis,s=y_axis)
```

```python
#creating a pie chart using a MATLAB method
plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,explode=explode_values,colors=colors,labels=x_axis,autopct='%.1f%%')
plt.show()
```

```python
#Creating a pie chart using object oriented interface
fig,ax = plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis,explode=explode_values,colors=colors,autopct='%.1f%%')
plt.show()
```
